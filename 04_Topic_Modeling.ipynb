We count the number of words in the documents in order to find the most recurrent words (that are not useful)

```python
#let's see if lemmatization improve the performances of the model

from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')

# definition of the object for lemmatization
wordnet_lemmatizer = WordNetLemmatizer()

# definition of the function for the lemmatization
def lemmatizer(text):
    words = text.split()  # string to list of words
    lemm_text = [wordnet_lemmatizer.lemmatize(word, pos='v') for word in words]
    lemmatized_string = ' '.join(lemm_text)
    return lemmatized_string


df_reviews_shuffled['review_text'] = df_reviews_shuffled['review_text'].apply(lambda x: lemmatizer(x))
df_reviews_shuffled.head()
```

```python
from collections import Counter

total_counter = Counter()
total_counter_words = Counter()

for index, row in df_reviews_shuffled.iterrows():
    count = Counter(row['review_text'].split())
    count_words = Counter(set(row['review_text'].split()))
    total_counter.update(count)
    total_counter_words.update(count_words)

print(total_counter)
print(total_counter_words)
```

```python
selected_words = [word for word, count in total_counter_words.items() if count > 10000]
print(selected_words)

def remove_short_words(text):
    words = text.split()
    filtered_words = [word for word in words if len(word) >= 3]
    return ' '.join(filtered_words)

def remove_words(text):
    text = ' '.join([word for word in text.split() if word.lower() not in selected_words])
    return text

docs = df_reviews_shuffled['review_text'][:100000]

docs = docs.apply(remove_words)
docs = docs.apply(remove_short_words)
```

## Function for coherence values

```python
from gensim.models import LdaModel
from gensim.models.coherencemodel import CoherenceModel
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer

def compute_coherence_values(dictionary, corpus, texts, limit, start=2, step=3, random_state=0):
    """
    Compute c_v coherence for various numbers of topics using LatentDirichletAllocation from scikit-learn

    Parameters:
    ----------
    dictionary : Gensim dictionary
    corpus : Gensim corpus
    texts : List of input texts
    limit : Max num of topics

    Returns:
    -------
    model_list : List of LDA topic models
    coherence_values : Coherence values corresponding to the LDA model with respective number of topics
    """
    coherence_values = []
    model_list = []

    for num_topics in range(start, limit, step):
        # Convert the Gensim corpus to a bag-of-words representation compatible with scikit-learn
        vectorizer = CountVectorizer()
        X = vectorizer.fit_transform([' '.join(doc) for doc in texts])

        # Fit LatentDirichletAllocation model
        lda = LatentDirichletAllocation(n_components=num_topics, random_state=random_state)
        lda.fit(X)

        # Get the document-topic matrix
        doc_topic_matrix = lda.transform(X)

        # Get the top words for each topic
        top_words_per_topic = []
        feature_names = vectorizer.get_feature_names_out()
        for topic_weights in lda.components_:
            top_words = [feature_names[i] for i in topic_weights.argsort()[:-10 - 1:-1]]  # Adjust the number of top words
            top_words_per_topic.append(top_words)

        # Calculate coherence using gensim's CoherenceModel
        coherencemodel = CoherenceModel(topics=top_words_per_topic, texts=texts, dictionary=dictionary, coherence='c_v')
        coherence_values.append(coherencemodel.get_coherence())
        model_list.append(lda)

    return model_list, coherence_values
```

## Function for coherence values

```python
def compute_coherence_values(doc, dictionary, corpus, texts, limit, start=2, step=1, random_state=0):
    """
    Compute c_v coherence for various numbers of topics using BERTopic

    Parameters:
    ----------
    dictionary : Gensim dictionary
    corpus : Gensim corpus
    texts : List of input texts
    limit : Max num of topics

    Returns:
    -------
    model_list : List of Bert topic models
    coherence_values : Coherence values corresponding to the LDA model with respective number of topics
    """
    coherence_values = []
    model_list = []

    for num_topics in range(start, limit, step):

        umap_model = UMAP.UMAP(n_neighbors=15,
                  n_components=10,
                  min_dist=0.0,
                  metric='cosine',
                  random_state=100)
        # Fit Bert model
        topic_model = BERTopic(umap_model=umap_model, language="english", calculate_probabilities=True, nr_topics=num_topics)
        topics, probabilities = topic_model.fit_transform(doc)
        topic_words = [[words for words, _ in topic_model.get_topic(topic)] for topic in range(len(set(topics))-1)]

        # Evaluate
        coherence_model = CoherenceModel(topics=topic_words,
                                 texts=texts,
                                 dictionary=dictionary,
                                 coherence='c_v')
        coherence_values.append(coherence_model.get_coherence())
        model_list.append(topic_model)
        print(num_topics)
    return model_list, coherence_values
```
